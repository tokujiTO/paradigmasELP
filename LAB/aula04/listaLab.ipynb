```python
# A)
import pandas as pd

data = {
    'Área de Negócio': ['Segurança', 'Financeiro', 'Transporte', 'Tecnologia', 'Marketing'],
    'Filial A': [7.8, 6.7, 2.8, 5.6, 8.3],
    'Filial B': [6.2, 4.1, 9.2, 4.5, 7.5],
    'Filial C': [9.8, 7.3, 3.4, 7.2, 9.1]
}

df = pd.DataFrame(data)

df.set_index("Área de Negócio", inplace=True)

print(df)
```

```python
# B)
df["Média Desempenho"] = df[['Filial A', 'Filial B', 'Filial C']].mean(axis=1).round(2)
df["Melhor Desempenho"] = df[['Filial A', 'Filial B', 'Filial C']].idxmax(axis=1)
df["Soma Desemepnho"] = df[['Filial A', 'Filial B', 'Filial C']].sum(axis=1)

print(df)
```

```python
# C)
df.to_csv("desempenho_filial.csv")

novo_dataFrame = pd.read_csv("desempenho_filial.csv", index_col=0)
print(novo_dataFrame)
```

```python

# EXERCÍCIO 2
import pandas as pd

# A)
df = pd.read_excel("pedidos_de_asilo.xlsx")

print("Dimensões do DataFrame: ", df.shape)
print(df.head())
```

```python
# B)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.hist(df['Número de Pedidos'], bins=100, color='blue', edgecolor='black')

plt.title('Histograma do Número de Pedidos de Asilo')
plt.xlabel('Número de Pedidos')
plt.xticks(range(0, 101, 10))
plt.xlim(0, 100)
plt.ylabel('Frequência')
plt.grid(True, linestyle='--', alpha=0.7)

```

```python
# C) grafico de dispersão
plt.figure(figsize=(10, 5))
plt.scatter(df['Ano'], df['Número de Pedidos'], color='red', alpha=0.5)
plt.title('Número de Pedidos de Asilo ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Número de Pedidos')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
